```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
import pandas as pd
import tensorflow as tf
import numpy as np
import math
```

```python

#  Import and read the train_modified.csv.
import pandas as pd
sales_predict_df = pd.read_csv("../Alex's Model/Resources/train_modified.csv")
sales_predict_df
```

```python
#Drop the target column "Item_Outlet_Sales" fro the feature DataFrame
features_df = sales_predict_df.drop(columns=['Item_Outlet_Sales','Item_Identifier','Outlet_Identifier','Item_Weight'])


```

```python
# # One-hot encoding using pd.get_dummies
# features_df = pd.get_dummies(features_df, columns=['Outlet_Size' ,'Item_Type' ,'Outlet_Location_Type','Outlet_Type','Outlet_Establishment_Year','Item_Fat_Content','Item_Weight'])

# # Extract target variables
target_df = sales_predict_df['Item_Outlet_Sales']
features_df.info()
```

```python

X = features_df.values
y = target_df.values
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape[1]
```

```python

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train_scaled).shape[1]),  # Input layer
   
      # Hidden layer 3
    
    tf.keras.layers.Dense(12, activation='relu'),
    tf.keras.layers.Dense(3, activation='relu'),
    tf.keras.layers.Dense(1)                           # Output layer
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=5, validation_split=.1, verbose=1)
```

```python
model_loss= model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}")

y_predict = model.predict(X_test_scaled)

r_squared = r2_score(y_test, y_predict)
print(f'R Square Score: {r_squared}')
```
